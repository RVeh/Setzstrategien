## Exakte Wahrscheinlichkeiten beim  ein-Würfelspiel (2 Personen)
Das folgende Programm wurde von ChatGPT 1:1 auf Anweisung (möglichst dicht an der Formel) aus dem Artikel übersetzt.

Es kann deutlich kompakter dargestellt werden. Das wurde auch auf Anweisung durchgeführt.

```python
from fractions import Fraction

# Wahrscheinlichkeiten p für die Felder als Brüche
p = [1/6, Fraction(10, 36), Fraction(8, 36), 
     Fraction(6, 36), Fraction(4, 36), Fraction(2, 36)]

# Speicher (Memoization) für alle drei Wahrscheinlichkeiten
memo = {}

# Gemeinsame Funktion zur Berechnung von P_A, P_B und P_U
def P_all(V, W):
    # Umwandlung der Listen in Tupel, um sie als Schlüssel im Memo zu verwenden
    key = (tuple(V), tuple(W))

    # Überprüfen, ob das Ergebnis schon berechnet wurde
    if key in memo:
        return memo[key]

    # Abbruchbedingungen
    if V == W:
        return (0, 0, 1)  # Unentschieden
    if all(x == 0 for x in V):
        return (1, 0, 0)  # A gewinnt
    if all(x == 0 for x in W):
        return (0, 1, 0)  # B gewinnt

    # Berechnung von s
    s = 1 - sum(p[j] for j in range(len(p)) if V[j] == 0 and W[j] == 0)

    # Rekursive Berechnung
    P_A_total = Fraction(0); P_B_total = Fraction(0); P_U_total = Fraction(0)

    for j in range(len(p)):
        # Initialisierung der nächsten Wahrscheinlichkeiten
        P_A_next = P_B_next = P_U_next = Fraction(0)

        if V[j] > 0 and W[j] > 0:
            Vj = V.copy()
            Wj = W.copy()
            Vj[j] -= 1
            Wj[j] -= 1
            P_A_next, P_B_next, P_U_next = P_all(Vj, Wj)
        elif V[j] == 0 and W[j] > 0:
            Wj = W.copy()
            Wj[j] -= 1
            P_A_next, P_B_next, P_U_next = P_all(V, Wj)
        elif V[j] > 0 and W[j] == 0:
            Vj = V.copy()
            Vj[j] -= 1
            P_A_next, P_B_next, P_U_next = P_all(Vj, W)

        # Berechnung der Gesamtergebnisse
        P_A_total += (p[j] / s) * P_A_next
        P_B_total += (p[j] / s) * P_B_next
        P_U_total += (p[j] / s) * P_U_next

    # Ergebnis speichern
    memo[key] = (P_A_total, P_B_total, P_U_total)
    return memo[key]

# Beispielzustände V und W
V = [3, 5, 4, 3, 2, 1]
W = [3, 6, 5, 3, 1, 0]

# Berechnung der Wahrscheinlichkeiten für A, B und Unentschieden
prob_A_wins, prob_B_wins, prob_U = P_all(V, W)

# Ausgabe der Ergebnisse
print(f"P_A({V},{W}) =  {float(prob_A_wins):5}")
print()
print(f"P_B({V},{W}) =  {float(prob_B_wins):5}")
print()
print(f"P_U({V},{W}) =  {float(prob_U):5}")
```

```python
from fractions import Fraction

# Wahrscheinlichkeiten p für die Felder als Brüche
p = [Fraction(6, 36), Fraction(10, 36), Fraction(8, 36), 
     Fraction(6, 36), Fraction(4, 36), Fraction(2, 36)]

# Speicher (Memoization) für alle drei Wahrscheinlichkeiten
memo = {}

# Gemeinsame Funktion zur Berechnung von P_A, P_B und P_U
def g(A, B, memo={}):
    # Memoization check
    if (tuple(A), tuple(B)) in memo:
        return memo[(tuple(A), tuple(B))]

    # Basisfälle
    if A == B:
        return (Fraction(0), Fraction(0), Fraction(1))  # Unentschieden
    if all(x == 0 for x in A):
        return (Fraction(1), Fraction(0), Fraction(0))  # A gewinnt
    if all(x == 0 for x in B):
        return (Fraction(0), Fraction(1), Fraction(0))  # B gewinnt

    # d_i berechnen
    d = [Fraction(0)] * 6
    for i in range(6):
        if A[i] > 0 or B[i] > 0:
            d[i] = p[i]

    # Summe s
    s = sum(d)

    # Rekursive Berechnung
    P_A, P_B, P_U = Fraction(0), Fraction(0), Fraction(0)

    for i in range(6):
        if A[i] > 0 or B[i] > 0:
            A_new = A.copy()  # Flache Kopie von A
            B_new = B.copy()  # Flache Kopie von B
            if A[i] > 0:
                A_new[i] -= 1
            if B[i] > 0:
                B_new[i] -= 1
            pa, pb, pu = g(A_new, B_new, memo)
            P_A += d[i] / s * pa
            P_B += d[i] / s * pb
            P_U += d[i] / s * pu

    # Memoization store
    memo[(tuple(A), tuple(B))] = (P_A, P_B, P_U)
    return P_A, P_B, P_U

# Generierung aller Setzstrategien mit n Chips und min/max Eingrenzungen
def generate_strategies(n, m, min_chips, max_chips):
    strategies = []
    
    # Hilfsfunktion zur Rekursion
    def generate(current, remaining, field):
        if field == m:
            if remaining == 0:
                strategies.append(current.copy())
            return
        
        # Versuche Chips auf das aktuelle Feld zu setzen, nur innerhalb der Grenzen
        for i in range(min_chips[field], min(max_chips[field], remaining) + 1):
            current[field] = i
            generate(current, remaining - i, field + 1)

    generate([0] * m, n, 0)
    return strategies

# Funktion, die zählt und Strategien ausgibt, gegen die eine bestimmte Strategie schlechter abschneidet
def count_losses_and_better_strategies(target_strategy, strategies):
    losses = 0
    better_strategies = []
    
    for other_strategy in strategies:
        if target_strategy == other_strategy:
            continue
        P_A, P_B, _ = g(target_strategy, other_strategy)
        if P_B > P_A:
            losses += 1
            better_strategies.append(other_strategy)  # Füge die Strategie zur Liste hinzu
    
    return losses, better_strategies

# Beispiel: Generiere Strategien für 18 Chips und 6 Felder mit min/max Grenzen
n = 18  # Anzahl der Chips
m = 6   # Anzahl der Felder

# Eingrenzungen der Chips auf den Feldern
min_chips = [2, 5, 2, 2, 1, 0]  # Minimale Anzahl der Chips auf jedem Feld
max_chips = [5, 8, 7, 5, 3, 1]  # Maximale Anzahl der Chips auf jedem Feld

# Generiere alle gültigen Strategien innerhalb der Grenzen
strategies = generate_strategies(n, m, min_chips, max_chips)

# Drei vorgegebene Strategien
#strategy_1 = [3, 6, 5, 3, 1, 0]  # Erste Strategie
strategy_1 =  [3, 6, 5, 2, 2, 0] # Erste Strategie
#strategy_2 = [3, 5, 4, 3, 2, 1]  # Zweite Strategie
#strategy_3 = [3, 7, 4, 3, 1, 0]  # Dritte Strategie

# Anzahl der Strategien und Liste der besseren Strategien für jede Strategie
losses_1, better_strategies_1 = count_losses_and_better_strategies(strategy_1, strategies)
#losses_2, better_strategies_2 = count_losses_and_better_strategies(strategy_2, strategies)
#losses_3, better_strategies_3 = count_losses_and_better_strategies(strategy_3, strategies)

# Ausgabe der Ergebnisse
print(f"Strategie {strategy_1} verliert gegen {losses_1} andere Strategien.")
print(f"Bessere Strategien: {better_strategies_1}")
print()
#print(f"Strategie {strategy_2} verliert gegen {losses_2} andere Strategien.")
#print(f"Bessere Strategien: {better_strategies_2}")
print()
#print(f"Strategie {strategy_3} verliert gegen {losses_3} andere Strategien.")
#print(f"Bessere Strategien: {better_strategies_3}")
```

## Wahrscheinlichkeiten (exakt) für 2 Würfel mithilfe der Rekursionsformel aus dem Artikel

Es können auch die Wahrscheinlichkeiten der Würfel getrennt definiert werden (z.B.: ein Würfel ist gezinkt)

```python
def compute_probabilities_two_dice(v, w, p_A, p_B, memo={}):
     # Convert list of tuples (v, w) to a hashable form for memoization
    state = (tuple(v), tuple(w))
    
    # Check if result for this state is already computed
    if state in memo:
        return memo[state]
    
    # Base cases:
    if v == [0] * len(v) and w != [0] * len(w):  # A has no chips, B still has chips -> A wins
        return 1.0, 0.0, 0.0  # A wins
    if w == [0] * len(w) and v != [0] * len(v):  # B has no chips, A still has chips -> B wins
        return 0.0, 1.0, 0.0  # B wins
    if v == [0] * len(v) and w == [0] * len(w):  # Both have no chips -> Draw
        return 0.0, 0.0, 1.0  # Draw

    # Calculate the normalization factor s
    s = sum(p_A[i] * p_B[j] for i in range(len(p_A)) for j in range(len(p_B)) if v[i] == 0 and w[j] == 0)

    # Initialize total probabilities for A winning, B winning, and a draw
    P_A_total, P_B_total, P_U_total = 0.0, 0.0, 0.0
    
    # Iterate over each combination of fields for A and B
    for i in range(len(v)):
        for j in range(len(w)):
            if v[i] > 0 and w[j] > 0:  # Both A and B have chips on these fields
                v_new = v.copy()
                w_new = w.copy()
                v_new[i] -= 1
                w_new[j] -= 1
                P_A, P_B, P_U = compute_probabilities_two_dice(v_new, w_new, p_A, p_B, memo)
                P_A_total += p_A[i] * p_B[j] * P_A
                P_B_total += p_A[i] * p_B[j] * P_B
                P_U_total += p_A[i] * p_B[j] * P_U
            
            elif v[i] > 0 and w[j] == 0:  # Only A has chips on field i
                v_new = v.copy()
                v_new[i] -= 1
                P_A, P_B, P_U = compute_probabilities_two_dice(v_new, w, p_A, p_B, memo)
                P_A_total += p_A[i] * p_B[j] * P_A
                P_B_total += p_A[i] * p_B[j] * P_B
                P_U_total += p_A[i] * p_B[j] * P_U
            
            elif v[i] == 0 and w[j] > 0:  # Only B has chips on field j
                w_new = w.copy()
                w_new[j] -= 1
                P_A, P_B, P_U = compute_probabilities_two_dice(v, w_new, p_A, p_B, memo)
                P_A_total += p_A[i] * p_B[j] * P_A
                P_B_total += p_A[i] * p_B[j] * P_B
                P_U_total += p_A[i] * p_B[j] * P_U
    
    # Normalize the probabilities by (1 - s)
    if s != 1:
        P_A_total /= (1 - s)
        P_B_total /= (1 - s)
        P_U_total /= (1 - s)
    
    # Memoize and return the result
    memo[state] = (P_A_total, P_B_total, P_U_total)
    return P_A_total, P_B_total, P_U_total

# Example setup with float for A and B having their own dice probabilities
v = [3, 5, 4, 3, 2, 1]  
w = [3, 6, 5, 3, 1, 0]  
p_A = [3/18, 5/18, 4/18, 3/18, 2/18, 1/18]  # Probabilities for A's dice
p_B = [3/18, 5/18, 4/18, 3/18, 2/18, 1/18]  # Probabilities for B's dice

# Calculate the probabilities that A wins, B wins, and a draw
P_A, P_B, P_U = compute_probabilities_two_dice(v, w, p_A, p_B)
print(f"P_A ({v},{w}) = {P_A}")
print(f"P_B ({v},{w}) = {P_B}")
print(f"P_U ({v},{w}) = {P_U}")
```
